```python
def round_to_5x(num):
    nearest_multiple=5*round(num/5) # note here is the process to downward round stuff
```

```python
def get_next_week_price(date,stock_price):
    d1=datetime.strptime(date,'%Y-%m-%d')
    global valid_days_between
    for delta in valid_days_between:
        d2=d1+timedelta(days=delta)
        d2_time_str=datetime.strftime(d2,%Y-%m-%d)
        if d2_time_str in stock_price:
                return stock_price[d2_time_str]
    return None
```
